```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with Aiml_unofficial.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}(?:\s(?:AM|PM))?\s-\s'
messages=re.split(pattern,data)[1:]

```

```python
print(messages)
```

```python
dates=re.findall(pattern,data)
```

```python
print(dates)
```

```python
df = pd.DataFrame({
    'user_messages': messages,
    'message_dates': dates
})


df['message_dates'] = df['message_dates'].str.replace(' -', '', regex=False)



df['dates'] = pd.to_datetime(
    df['message_dates'],
    format='mixed',
    dayfirst=True,
    errors='coerce'
)


df.drop(columns='message_dates', inplace=True)
```

```python

df.head()
```

```python
df.shape
```

```python
#seprate user and messages
users = []
messages = []

for i in df['user_messages']:
    entry = re.split(r'^([^:]+):\s', i, maxsplit=1)
    
    if len(entry) == 3:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['messages'] = messages


df.drop(columns='user_messages',inplace=True)
```
